# O que é o Bagging:

O bagging é uma técnica que combina múltiplos modelos de machine learning para melhorar o desempenho preditivo. Fazendo com que o resultado previsto seja a média dos valores previstos por cada modelo

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.utils import resample
```

```python
df = pd.read_excel('/content/vendas.xlsx')
```

```python
df.head(2)
```

```python
df = pd.get_dummies(df)
```

```python
modelos = []
```

```python
for i in range(5):
  amostra = resample(df, replace=True, random_state=i)
  X = amostra.drop('quantidade_de_produtos', axis=1)
  y = amostra['quantidade_de_produtos']

  modelo = LinearRegression()
  modelo.fit(X, y)

  modelos.append(modelo)
```

```python
X_teste = df.drop('quantidade_de_produtos', axis=1)
y_teste = df['quantidade_de_produtos']
```

```python
previsoes = [modelo.predict(X_teste) for modelo in modelos]
```

```python
previsao_final = pd.DataFrame(previsoes).mean(axis=0)
```
